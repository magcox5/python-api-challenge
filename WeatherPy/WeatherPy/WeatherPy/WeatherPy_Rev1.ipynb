```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import requests
import time
from datetime import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = list(zip(lats, lngs))

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Create dataframe to hold weather for city data
weather_df = pd.DataFrame(cities)
# Rename column
weather_df.rename(columns = {0:'City'}, inplace = True)
weather_df.head()
```

```python
# Add Columns for Temp, Humidity, Cloudiness, and Windspeed
weather_df['Country'] = ''
weather_df['Lat'] = ''
weather_df['Lon'] = ''
weather_df['Date'] = ''
weather_df['Temp (F)'] = ''
weather_df['Humidity (%)'] = ''
weather_df['Cloudiness (%)'] = ''
weather_df['Wind Speed (mph)'] = ''

weather_df.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

counter = 1
for index, row in weather_df.iterrows():
    
    city = row['City']
    target_url = f"{base_url}appid={api_key}&q={city}&units={units}"
    
    print(f"#{counter} - Getting data for {row['City']}")
    weather_response = requests.get(target_url).json()
    
    # Try getting all for data points for each city... some may contain no data
    try:
        weather_df.loc[index, "Lat"] = weather_response["coord"]["lat"]
        weather_df.loc[index, "Lon"] = weather_response["coord"]["lon"]
        weather_df.loc[index, "Date"] = weather_response["dt"]
        weather_df.loc[index, "Country"] = weather_response["sys"]["country"]
        weather_df.loc[index, "Temp (F)"] = weather_response["main"]["temp"]
        weather_df.loc[index,"Humidity (%)"] = weather_response["main"]["humidity"]
        weather_df.loc[index,"Cloudiness (%)"] = weather_response["clouds"]["all"]
        weather_df.loc[index,"Wind Speed (mph)"] = weather_response["wind"]["speed"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")

    counter += 1
    
weather_df.head()
```

```python
# Remove cities with no weather data
weather_df = weather_df[weather_df['Temp (F)'] != '']
weather_df.head()
```

#### Latitude vs. Temperature Plot

```python
# Set up constants to be used in plots
latitude = weather_df['Lat']
temp = weather_df['Temp (F)']
humidity = weather_df['Humidity (%)']
cloudiness = weather_df['Cloudiness (%)']
windspeed = weather_df['Wind Speed (mph)']
min_temp = weather_df['Temp (F)'].min()
max_temp = weather_df['Temp (F)'].max()
date_data = weather_df['Date'][0]
#dt = (datetime.fromtimestamp(date_data)).strftime('%Y-%m-%d %H:%M:%S')
dt = (datetime.fromtimestamp(date_data)).strftime('%m-%d-%Y')
print(f"The minimum temp is {min_temp} and the maximum temp is {max_temp}")
print(dt)
```

```python
# Latitude vs. Temperature Plot
plt.scatter(latitude, temp, marker="o", facecolors="red", edgecolors="black")
# Create a title, x label, and y label for our chart
plt.title(f"Latitude vs. Temperature\n Date: {dt}")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.xlim(-90,90)
plt.ylim(min_temp-10, max_temp+10)
plt.margins(.2)
plt.savefig('latitude_vs_temp.png')
plt.show()
```

#### Latitude vs. Humidity Plot

```python
# Latitude vs. Humidity Plot
plt.scatter(latitude, humidity, marker="o", facecolors="red", edgecolors="black")
# Create a title, x label, and y label for our chart
plt.title(f"Latitude vs. Humidity\n Date: {dt}")
plt.xlabel("Latitude")
plt.ylabel("Humidity %")

plt.savefig('latitude_vs_humidity.png')
plt.show()
```

```python
# Latitude vs. Cloudiness
plt.scatter(latitude, cloudiness, marker="o", facecolors="red", edgecolors="black")
# Create a title, x label, and y label for our chart
plt.title(f"Latitude vs. Cloudiness %\n Date: {dt}")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness %")

plt.savefig('latitude_vs_clouds.png')
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# Latitude vs. Wind Speed
plt.scatter(latitude, windspeed, marker="o", facecolors="red", edgecolors="black")
# Create a title, x label, and y label for our chart
plt.title(f"Latitude vs. Windspeed\n Date: {dt}")
plt.xlabel("Latitude")
plt.ylabel("Windspeed (mph)")

plt.savefig('latitude_vs_wind.png')
plt.show()
```
